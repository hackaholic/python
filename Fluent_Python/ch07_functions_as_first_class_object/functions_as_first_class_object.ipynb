### Treating a Function like an object

```python
def factorial(n):
    """ return n!"""
    return 1 if n < 2 else n * factorial(n-1)

factorial(5)
```

### User-Defined Callable types

```python
import random
class BingoCage:
    def __init__(self, items):
        self._items = list(items)
        random.shuffle(self._items)
    
    def pick(self):
        try:
            return self._items.pop()
        except IndexError:
            raise LookupError("pick from empty bingoCage")
    
    def __call__(self):
        return self.pick()
```

```python
bingo = BingoCage(range(3))
bingo()
```

```python
callable(bingo)
```

### The operator module

```python
from functools import reduce
from operator import mul
from operator import itemgetter, attrgetter, methodcaller
```

```python
def factorial1(n):
    return reduce(lambda x,y: x*y, range(1, n+1))
```

```python
def factorial2(n):
    return reduce(mul, range(1, n+1))
```

```python
metro_data = [
    ("Tokyo", "JP", 36.933, (35.689722, 139.691667)),
    ("Delhi NCR", "IN", 21.935, (28.613889, 77.208889)),
    ("Mexico City", "MX", 20.142, (19.433333, -99.133333)),
    ("New York-Newark", "US", 20.104, (40.808611, -74.020386)),
    ("São Paulo", "BR", 19.649, (-23.547778, -46.635833)),
]
```

```python
for city in sorted(metro_data, key=itemgetter(1)):
    print(city)
```

```python
cc_name = itemgetter(1,0)
for city in metro_data:
    print(cc_name(city))
```

A sibling of itemgetter is attrgetter which creates function to extract object attributes by name

```python
from collections import namedtuple
```

```python
LatLon = namedtuple('LatLon', 'lat lon')
Metropolis = namedtuple('Metropolis', ['name', 'cc', 'prop', 'coord'])

metro_areas = [ Metropolis(name, cc , prop, LatLon(lat, lon)) for name, cc, prop, (lat, lon) in metro_data]
metro_areas
```

```python
sorted(metro_areas, key=attrgetter('coord.lat'))
```
