# Decorator without argument

```python
class test:
    def __init__(self, f):
        self.f = f
        print("Init  class")
    def __call__(self):
        print(f"I am the inside call {self.f.__name__}")
        self.f()
```

```python
@test
def hello():
    print("I am inside hello")
```

```python
hello()           # when the function is called __call__ is invoked first
```

# function having argument

```python
class test:
    def __init__(self, f):
        self.f = f
        print("Init class")
    def __call__(self, a, b):
        print("I am inside call")
        self.f(a,b)
```

```python
@test
def hello(a,b):
    print(f"I am inside hello: {a}, {b}")
```

```python
hello(1, 2)
```

# Decorator and function both have argument

```python
class test:
    def __init__(self, arg1, arg2):
        self.arg1 = arg1
        self.arg2 = arg2
        print(f"Print decorator argument: {self.arg1}, {self.arg2}")
    def __call__(self, f):
        self.f = f
        def inner(a,b):
            print(f"I am inner recieved arg: {a}, {b}")
            self.f(a,b)
        return inner
```

```python
@test("Done", "Dtwo")
def hello(a, b):
    print(f"I am inside hello {a}, {b}")
```

```python
hello(1,2)
```

# Divide by zero example

```python
class dividebyzero:
    def __init__(self, f):
        self.f = f
    def __call__(self, a, b):
        if b==0:
            print("Divide by zero Error")
            #raise ZeroDivisionError
        else:
            return self.f(a,b)

@dividebyzero
def divide(a,b):
    return a/b
```

```python
k = divide(1,0)
print(k)
```

# Multiple Decorator

```python
class Decorator1:
    def __init__(self, func):
        self.func = func
    
    def __call__(self, *args, **kwargs):
        print("Decorator 1 - Before function call")
        result = self.func(*args, **kwargs)
        print("Decorator 1 - After function call")
        return result

class Decorator2:
    def __init__(self, func):
        self.func = func
    
    def __call__(self, *args, **kwargs):
        print("Decorator 2 - Before function call")
        result = self.func(*args, **kwargs)
        print("Decorator 2 - After function call")
        return result

@Decorator1
@Decorator2
def my_function(x, y):
    return x + y
```

```python
result = my_function(3, 5)
print("Final Result:", result)
```
